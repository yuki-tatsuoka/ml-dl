```python
#問題１
#紙の厚さ
THICKNESS = 0.00008

#この紙を43回折ってみる。ただし、折った後のサイズは2の43乗とする。
folded_thickness = THICKNESS*2**43
print("厚さ： {}メートル".format(folded_thickness))


#問題2
#上記のメートルをキロメートルに変換する
print("厚さ： {:.2f}キロメートル".format(folded_thickness/1000))


#問題3
#for文を使用して作成する
for i in range(44):
    power = 2 ** i
    folded_thickness  = THICKNESS * power

print("厚さ： {:.2f}キロメートル".format(folded_thickness/1000))
```

```python
#問題4
#コードの処理速度を把握する。
import time
start = time.time()

#print文の速度比較
folded_thickness = THICKNESS*2**43

elapses_time = time.time() - start
print("time: {:.8f}[s]".format(elapses_time))

#for文の比較
start = time.time()
for i in range(44):
    power = 2 ** i
    folded_thickness  = THICKNESS * power
elapses_time = time.time() - start
print("time: {:.8f}[s]".format(elapses_time))
```

```python
%%timeit 

#問題４の続き。%%timeitでも計測してみる。
for i in range(44):
    power = i ** 2
    folded_thickness  = THICKNESS * power
```

```python
#問題５
#for文で作成したコードをリストへ格納する
sample_list = []
for i in range(44):
    power = 2 ** i
    sample_list.append(power)
len(sample_list)
```

```python
#問題6
#グラフを表示する
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(0)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(sample_list)
plt.show()

#考察
#グラフの前半は緩やかに増加しているが、後半の４０手前あたりから指数的に増加している。

#問題7
#グラフのカスタマイズ。３つカスタマイズして記載する。
#図形1 labelの色を赤にしてサイズを変更する
plt.figure(1)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(sample_list)
plt.tick_params(color="red",labelsize=15,labelcolor="red")
plt.show()

#図形2 labelの最大値と最小値を変更する
import numpy as np
plt.figure(2)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.xlim(0,len(sample_list)+5)#最大値に５増やして幅を作る
plt.ylim(np.min(sample_list),np.max(sample_list))
plt.plot(sample_list)
plt.show()

#図形3 色を青に、太さを変更する
plt.figure(3)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(sample_list, color="blue", lw=3, marker="o")
plt.show()
```
